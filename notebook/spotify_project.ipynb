- Both streams and in_deezer_playlists are object data types when I'd expect them to be integers
    - Will need to adjust these rows/values so they can be properly aggregated

```python
# Converting streams to floats and adding error handling for values that cannot be converted

df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df[df['streams'].isna() == True] # Identifying row where dtype conversion does not work

df['streams'].dtypes # Validating the streams columns is the expected data type
```

```python
df[df['in_deezer_playlists'].str.contains(',') == True] # Identifying rows with commas in the in_deezer_playlists column
df['in_deezer_playlists'] = df['in_deezer_playlists'].str.replace(',', '') # Replacing commas with nothing
df[df['in_deezer_playlists'].str.contains(',') == True] # Validating replacement worked


```

```python
# Converting in_deezer_playlists to floats

df['in_deezer_playlists'] = pd.to_numeric(df['in_deezer_playlists'], errors='coerce')

df.describe()
```

```python
# Who are the top 10 artists that appear in the most streamed dataset? How do their song characteristics look compared to the rest of the dataset?

top_10 = df.value_counts('artist(s)_name').head(10)
top_10_mask = df['artist(s)_name'].isin(top_10.index)
remaining_mask = ~top_10_mask

top_10
```

```python
df[top_10_mask][['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].mean()
```

```python
df[remaining_mask][['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].mean()
```

**Top 10 Artists:** 
- Taylor Swift is the clear leader in most streamed songs showing up 34 times in the dataset
- All of the top 10 artists show up in the dataset multiple times, suggesting streaming success across multiple releases
- Looking at the average song characteristics between the Top 10 artists and the remaining dataset, there isn't a clear feature that separates the group at the moment
    - This may be down to different genres being averaged together or the Remaining group having a larger selection (and potentially variety) of songs

```python
# How do top streamed songs trend over time?

by_year = df.groupby('released_year')['track_name'].count().sort_values(ascending=True)
plt.figure(figsize=(10, 6))
by_year.plot(kind='bar')
plt.title('Top Streamed Songs by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Song Count')
plt.show()
```

- There is a higher concentration of top streaming sings in the 2000s, which might make sense given that music streaming apps have become more relevant over the last decade. This isn't to say songs released prior to the 2000s cannot be streamed, but 

```python
# What key has the most songs written in? What are the characteristics of the songs in this group and how does it compare to the remainder of the dataset?

top_key = df.groupby('key')['track_name'].count().sort_values(ascending=False).head(1)
top_key_mask = df['key'].isin(top_key.index)
remainder_mask = ~df['key'].isin(top_key.index)

top_key
```

```python
top = df[top_key_mask][['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].mean()
top.round(2)
```

```python
remaining = df[remainder_mask][['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].mean()
remaining.round(2)
```

**Song Key Details:**
- The most popular key for songs that were streamed was C#
- Though not significantly different, songs in C# were slightly more energetic and more speechy. Acousticness and instrumentalness were also lower, suggesting listener preference for a more natural sound

**Streams vs. Charts:**
- Based on the scatterplot, there aren't any clear visual patterns that suggest being present in more charts will lead to more streams
    - Logically speaking, it would also likely matter what specific charts a song was in along with how wide reaching that chart is
        - A song in Top 50 USA might have a different reach/audience compared to Top 50 Argentina

**Streams vs. Playlists:**
- While streams and charts showed no correleation, the opposite holds true for streams vs. playlists. Across all three platforms in the dataset, the more playlists a song was a part of, the more streams it saw
- The Spotify and Apple scatterplots capture this dynamic the best
- Because charts are more objective and factor in multiple datapoints to determine a song's popularity, they do not offer the same level of personalization as playlists do
    - Users can create playlists tailored to their lifestyle or mood and play their favorite songs more regularly
    - For songs that are high in charts, they may not have the same level of visibility once they fall off whatever charts they were on
